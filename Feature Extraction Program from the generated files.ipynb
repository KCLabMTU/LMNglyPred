```python
import os 
import pandas as pd
import numpy as np

os.chdir("/home/t326h379/Prot_T5/Prot_T5_csv")

Header_name = ["Amino_Acid"]

col_of_feature = [i for i in range(1,1025)]

Header_name = Header_name + col_of_feature

def open_file(label_i, protein_id, csv_file_name, Position_individual,Sequences_individual):  

    
    df = pd.read_csv(csv_file_name, header=None)
    
    df.columns = Header_name

    
        

    with open(csv_file_name) as outfile:
        feature_vector = []
        AA = []
        lines = outfile.readlines()
        for line in lines:
            list = line.split(",")
            AA.append(list[0])
           
        AA = ''.join(AA)
        Position_individual = int(Position_individual)
        
        
        index = Position_individual-1
        
        if (AA[index] == "N"):
        

            matrix = df.iloc[index]
            grabbed_matrix =  np.array(matrix)


            flatten_value = grabbed_matrix.ravel()

            feature_vector = flatten_value.tolist()


            listToStr = ','.join([str(elem) for elem in feature_vector])
            



            with open('Feature_Extraction_August_12_Subash_123_DeepNGlyPred_data.txt','a+') as outfile:
                actual_index_when_positioned_from_one = Position_individual
                
                outfile.write(str(label_i))
                outfile.write(",")
                outfile.write(str(protein_id))
                outfile.write(",")
                outfile.write(str(actual_index_when_positioned_from_one))
                outfile.write(",")
                outfile.write(Sequences_individual)
                outfile.write(",")

                outfile.write(listToStr)
                outfile.write("\n")

PID = []
Position = []
Sequences = []
label = []
with open("N_linked_glycosylation_protein_training_file_without_independnet_testing_protein.txt") as hoverboard:
    for line in hoverboard:
        x = line.split()
        PID.append(x[1])
        Position.append(x[2])
        Sequences.append(x[3])
        label.append(x[0])
        
for i in range(len(Sequences)):
    
    print(i,"\t",PID[i],"\t",Position[i])
    protein_id = PID[i]
    label_i = label[i]
    csv_file_name = PID[i]+"_Prot_Trans_.csv"
    Position_individual = Position[i]
    Sequences_individual = Sequences[i].strip("\n").strip("-")
    open_file(label_i, protein_id, csv_file_name, Position_individual,Sequences_individual)
```
